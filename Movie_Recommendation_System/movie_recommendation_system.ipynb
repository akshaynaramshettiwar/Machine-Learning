```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
# checking null values 
# here we have only 3 null values and now we can drop them
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
# let's check if we have any duplicate values
movies.duplicated().sum()
```

```python
# let's mere some of the features
import ast
ast.literal_eval

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply([lambda x:[i.replace(' ','')for i in x]])
```

```python
movies['genres']
```

```python
movies['genres1']=movies['genres'].apply([lambda x:[i.replace(' ','')for i in x]])
```

```python
movies['cast']=movies['cast'].apply([lambda x:[i.replace(' ','')for i in x]])
movies['crew']=movies['crew'].apply([lambda x:[i.replace(' ','')for i in x]])
movies['keywords']=movies['keywords'].apply([lambda x:[i.replace(' ','')for i in x]])
```

```python
movies.head()
```

```python
movies.drop(['genres1'],axis=1,inplace=True)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new
```

```python
new['tags']=new['tags'].apply([lambda x:' '.join(x)])
```

```python
new['tags'][0]
```

```python
new['tags']=new['tags'].apply([lambda x: x.lower()])
```

```python
new['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
# from these most repeating words we can see there are similar words like africa,african etc 
# so we have to remove them
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
new['tags']
```

```python
def steam(text):
    ls=[]
    for i in text.split():
        ls.append(ps.stem(i))
        
    return " ".join(ls)
```

```python
new['tags']=new['tags'].apply(steam)
```

```python
new['tags']
```

```python
new['tags'][0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity.shape
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
index = new[new['title'] == 'Avatar'].index[0]
distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
for i in distances[1:6]:
    print(new.iloc[i[0]].title)
```
